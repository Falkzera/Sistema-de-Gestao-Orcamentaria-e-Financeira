```python
import pandas as pd
```

```python
df = pd.read_excel("ted.xlsx", sheet_name="TED")
```

```python
df.head()
```

```python
# remover MÊS, ANO

df = df.drop(columns=["MÊS", "ANO"])
df = df.rename(columns={"Valor Inicial": "Valor", "Data inicial": "Data de Publicação", "Data final": "Data de Encerramento"})
```

```python
df.head()
```

```python
# Data de Publicação -> strftime("%d/%m/%Y")
df["Data de Publicação"] = df["Data de Publicação"].dt.strftime("%Y-%m-%d")
df["Data de Encerramento"] = df["Data de Encerramento"].dt.strftime("%Y-%m-%d")
df["Data de Recebimento"] = df["Data de Recebimento"].dt.strftime("%Y-%m-%d")
```

```python
# trnasofmrar Valor em float, o que nao der deixar em vazio
def transform_value(value):
    try:
        return float(value)
    except (ValueError, TypeError):
        return None
    
df["Valor"] = df["Valor"].apply(transform_value)
```

```python
# verifica se há 2 valores apos o . e adiciona duas casas decimais se não houver ou so houver 1 digito
def format_value(value):
    if pd.isna(value):
        return value
    return f"{value:.2f}"

df["Valor"] = df["Valor"].apply(format_value)
```

```python
df.head()
```

```python
# salvar como base_filtrada xlsx
df.to_excel("base_filtrada.xlsx", index=False)
```
