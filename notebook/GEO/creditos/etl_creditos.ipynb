```python
import pandas as pd
```

```python
df = pd.read_excel("creditos.xlsx", sheet_name="Processos Base")
```

```python
df.head()
```

```python
df.columns
```

```python
# construir uma coluna denominada "Nº ATA", na coluna Observação, coletar as informações que tenham o inicio com "Ata", e colocar na coluna "Nº ATA" e remover essas informações da coluna "Observação"
df["Nº ATA"] = df["Observação"].str.extract(r"Ata (\d+)", expand=False)
df["Observação"] = df["Observação"].str.replace(r"Ata \d+", "", regex=True)
```

```python
# adicionar coluna Opnião SOP aopenas
df["Opnião SOP"] = df["Observação"].str.extract(r"Opnião SOP: (.+)", expand=False)
```

```python
df.head()
```

```python
# data de recebimento  format=ano mes dia')
df["Data de Recebimento"] = pd.to_datetime(df["Data de Recebimento"], format="%Y-%m-%d")
df["Data de Publicação"] = pd.to_datetime(df["Data de Publicação"], format="%Y-%m-%d")

# removendo os horariso
df["Data de Recebimento"] = df["Data de Recebimento"].dt.date
df["Data de Publicação"] = df["Data de Publicação"].dt.date
```

```python
df.head()
```

```python
df.to_excel("base_filtrada.xlsx", index=False)
```
