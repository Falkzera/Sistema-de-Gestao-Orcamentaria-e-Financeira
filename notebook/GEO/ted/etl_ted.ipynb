```python
import pandas as pd
```

```python
df = pd.read_excel("ted_att.xlsx", sheet_name="FALCAO")
```

```python
df.head()
```

```python
# data de publicação convertido para datetime, inicialmente está em foramto 45792 
df["Data de Publicação"] = pd.to_datetime(df["Data de Publicação"], unit="D", origin="1899-12-30")
df["Data de Encerramento"] = pd.to_datetime(df["Data de Encerramento"], unit="D", origin="1899-12-30")
df["Data de Recebimento"] = pd.to_datetime(df["Data de Recebimento"], unit="D", origin="1899-12-30")
```

```python
# Data de Publicação -> strftime("%d/%m/%Y")

df["Data de Publicação"] = pd.to_datetime(df["Data de Publicação"], errors="coerce").dt.strftime("%Y-%m-%d")
df["Data de Encerramento"] = pd.to_datetime(df["Data de Encerramento"], errors="coerce").dt.strftime("%Y-%m-%d")
df["Data de Recebimento"] = pd.to_datetime(df["Data de Recebimento"], errors="coerce").dt.strftime("%Y-%m-%d")
```

```python
df['Saldo'] = df['Valor Inicial'] - df['Valor Descentralizado']
```

```python
# trnasofmrar Valor em float, o que nao der deixar em vazio
def transform_value(value):
    try:
        return float(value)
    except (ValueError, TypeError):
        return None
    
df["Valor Inicial"] = df["Valor Inicial"].apply(transform_value)
df["Valor Descentralizado"] = df["Valor Descentralizado"].apply(transform_value)
df["Saldo"] = df["Saldo"].apply(transform_value)
```

```python
# verifica se há 2 valores apos o . e adiciona duas casas decimais se não houver ou so houver 1 digito
def format_value(value):
    if pd.isna(value):
        return value
    return f"{value:.2f}"

df["Valor Inicial"] = df["Valor Inicial"].apply(format_value)
df["Valor Descentralizado"] = df["Valor Descentralizado"].apply(format_value)
df["Saldo"] = df["Saldo"].apply(format_value)
```

```python
df = df[['Situação TED', 'Nº do Processo', 'UO Concedente', 'UO Executante', 'Nº do TED', 'Termo Aditivo', 'Programa de Trabalho', 'Fonte de Recursos', 'Natureza de Despesa', 'Objetivo', 'Valor Inicial', 'Valor Descentralizado', 'Saldo', 'Data de Publicação', 'Data de Encerramento', 'Data de Recebimento']]
```

```python
df.head()
```

```python
# Rename -> Situação -> Situação TED
df = df.rename(columns={"Situação": "Situação TED"})
```

```python
# para as colunas  UO Concedente e UO Executante, onde tiver PM -> PMAL
df['UO Concedente'] = df['UO Concedente'].str.replace("PM", "PMAL", regex=False)
df['UO Executante'] = df['UO Executante'].str.replace("PM", "PMAL", regex=False)
```

```python
# em natureaa de despesa, trocar ; -> ,
df['Natureza de Despesa'] = df['Natureza de Despesa'].str.replace(";", ",", regex=False)
```

```python
# coluna programa de trabalho -> se a linha contiver 13 digitos, deve ser posto deste modo -> 00.000.0000.0000, se houver 12 digitos -> adicionar um zero no final e executar o primeiro if , se nao deixa como está
def format_programa_trabalho(value):
    if pd.isna(value):
        return value
    parts = value.split(";")
    formatted_parts = []
    for part in parts:
        part = part.strip()
        if len(part) == 13 and part.isdigit():
            formatted_parts.append(f"{part[:2]}.{part[2:5]}.{part[5:9]}.{part[9:]}")
        elif len(part) == 12 and part.isdigit():
            part += "0"
            formatted_parts.append(f"{part[:2]}.{part[2:5]}.{part[5:9]}.{part[9:]}")
        else:
            formatted_parts.append(part)
    return "; ".join(formatted_parts)



df['Programa de Trabalho'] = df['Programa de Trabalho'].astype(str).apply(format_programa_trabalho)
```

```python
df['Programa de Trabalho'] = df['Programa de Trabalho'].str.replace(";", ",", regex=False)
df['Programa de Trabalho'] = df['Programa de Trabalho'].str.replace("nan", "", regex=False)
```

```python
df.head()
```

```python
# astype str Natureza de Despesa
df['Natureza de Despesa'] = df['Natureza de Despesa'].astype(str)
# replace nan -> ''
df['Natureza de Despesa'] = df['Natureza de Despesa'].str.replace("nan", "", regex=False)

```

```python
# rename Valor Inicial -> Valor
df = df.rename(columns={"Valor Inicial": "Valor"})
```

```python
df.head()
```

```python
# orgão_uo o que tiver PC -> PCAL
df['UO Concedente'] = df['UO Concedente'].str.replace("PC", "PCAL", regex=False)
```

```python
# salvar como base_filtrada xlsx
df.to_excel("base_filtrada.xlsx", index=False)
```
